```python
%%bash
city=Casablanca
```

```python
%%bash
curl -s wttr.in/$city?T --output weather_report
```

```python
%%bash
#To extract Current Temperature
obs_temp=$(curl -s wttr.in/$city?T | grep -m 1 '째.' | grep -Eo -e '-?[[:digit:]].*')
echo "The current Temperature of $city: $obs_temp"
```

```python
%%bash
# To extract the forecast tempearature for noon tomorrow
fc_temp=$(curl -s wttr.in/$city?T | head -23 | tail -1 | grep '째.' | cut -d 'C' -f2 | grep -Eo -e '-?[[:digit:]].*')
echo "The forecasted temperature for noon tomorrow for $city : $fc_temp C"
```

```python
%%bash
#Store the current hour, day, month, and year in corresponding shell variables
hour=$(TZ='Morocco/Casablanca' date -u +%H) 
day=$(TZ='Morocco/Casablanca' date -u +%d) 
month=$(TZ='Morocco/Casablanca' date +%m)
year=$(TZ='Morocco/Casablanca' date +%Y)
```

```python
%%bash
#Merge the fields into a tab-delimited record, corresponding to a single row in Table 1
record=$(echo -e "$year\t$month\t$day\t$obs_temp\t$fc_temp C")
echo $record>>rx_poc.log
```
